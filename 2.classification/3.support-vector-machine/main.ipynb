<h2>Support Vector Machine (SVM)</h2>
<h4>Description</h4>
Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks. In classification, SVM finds the best boundary (hyperplane) that separates different classes of data with the maximum margin. <br>

Support Vector Machine (SVM) is primarily used for binary classification, where it aims to separate data into two distinct classes.

$$
    y = \text{sign}(w * x_i + b)
$$

<h4>Update Rule (Gradient Descent & Regularization Term)</h4>
Cost function:

$$ 
    L = \lambda \lvert w \rvert^2 + \frac{1}{n} \sum_{i=1}^{n} \text{max}(0,1 - y_i(w * x_i - b))
$$


If classification was successfull then we use:
$$
    w = w - \alpha * \frac{\partial L_i}{\partial w} = w - \alpha * 2 \lambda w
$$

$$
    b = b - \alpha * \frac{\partial L_i}{\partial b} = b
$$

Otherwise:
$$
    w = w - \alpha * \frac{\partial L_i}{\partial w} = w - \alpha * (2 \lambda w - y_i x_i)
$$
$$
    b = b - \alpha * \frac{\partial L_i}{\partial b} = b - \alpha * y_i
$$

- $\alpha$: learning rate;
- $\lambda$: lambda parameter;
- w: weights (coefficients);
- b: bias (intercept);

<h4>Importing libraries</h4>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

<h4>Creating dataset</h4>

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=50, n_features=2, centers=2, cluster_std=1.05, random_state=40)
y = np.where(y == 0, -1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

<h4>Training the model</h4>

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train, y_train)
```

<h4>Checking the accuracy</h4>

```python
from sklearn.metrics import accuracy_score

y_pred = svc.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

<h4>Visualising results</h4>

```python
plt.scatter(X_test[:, 0], X_test[:, 1], marker="o", c=y_pred)
plt.show()
```
