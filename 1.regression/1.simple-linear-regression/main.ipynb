<h2>Simple Linear Regression (SLR)</h2>

<h4>Model description</h4>
SLR is a linear regression model with a single explanatory variable. That is, it concerns two-dimensional sample points with one independent variable and one dependent variable (conventionally, the x and y coordinates in a Cartesian coordinate system) 


$$Y = b_0 + b_1 X_1$$
- $Y$: dependent variable;
- $b_0$: y-intercept (constant);
- $b_1$: slope coefficient;
- $X_1$: independent variable;

<h4>Loading data set</h4>

```python
import pandas as pd

df = pd.read_csv("salary-data.csv")

print(df.isnull().sum())
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

<h4>Splitting data into training & testing datasets</h4>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

<h4>Training the SLR model on the training dataset</h4>

```python
from sklearn.linear_model import LinearRegression

slr_model = LinearRegression()
slr_model.fit(X_train, y_train)
```

<h4>Predicting the test results</h4>

```python
y_pred = slr_model.predict(X_test)
```

<h4>Visualising the training set results</h4>

```python
import matplotlib.pyplot as plt

plt.scatter(X_train, y_train, color="red")
plt.plot(X_train, slr_model.predict(X_train), color="blue")

plt.title("Salary vs Experience (Testing set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")

plt.show()
```

<h4>Visualising the testing set results</h4>

```python
plt.scatter(X_test, y_test, color="red")
plt.plot(X_test, y_pred, color="blue")

plt.title("Salary vs Experience (Training set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")

plt.show()
```

<h4>Getting the final linear regresssion equation with the values of the coefficients</h4>

```python
print(slr_model.intercept_)
print(slr_model.coef_)
```
