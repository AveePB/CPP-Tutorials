<h2>Support Vector Regression (SVR)</h2>

<h4>Introduction</h4>
In the context of machine learning, SVR stands for Support Vector Regression. It is a type of Support Vector Machine (SVM) used for regression tasks. SVR aims to predict continuous output values given input data, much like other regression algorithms.

$$ 𝑓(𝑥) = 𝑤 ^𝑇 𝑥 + 𝑏 $$
- 𝑥: is the vector of input features;
- 𝑤: is the vector of weights (coefficients) associated with each feature;
- 𝑏:  is the bias term (intercept);


<h4>How it works</h4>
The goal of SVR is to find a function 𝑓(𝑥) that approximates the relationship between input variables (features) and a continuous target variable as accurately as possible, while maintaining a margin of tolerance. <br>

SVR uses a different loss function than typical regression algorithms. It aims to keep the error within a specified threshold, known as the epsilon (𝜖) margin. Errors within this margin are not penalized, while errors outside the margin are penalized linearly. <br>

In SVR, the regression function $𝑓(𝑥) = 𝑤 ^𝑇 𝑥 + 𝑏$ represents a hyperplane in a high-dimensional space. The algorithm tries to position this hyperplane such that the maximum number of data points lie within the epsilon margin. <br>

Similar to SVM for classification, SVR relies on support vectors, which are the data points that lie outside the epsilon margin or exactly on the boundary. These points determine the position and orientation of the regression hyperplane.





<h4>Importing the dataset</h4>

```python
import pandas as pd

df = pd.read_csv("Position_salaries.csv")
X = df.iloc[:, 1].values.reshape(-1, 1)
y = df.iloc[:, -1].values.reshape(-1, 1)
```

```python
print(X)
```

```python
print(y)
```

<h4>Feature Scaling</h4>

```python
from sklearn.preprocessing import StandardScaler

X_sc = StandardScaler()
X = X_sc.fit_transform(X)

y_sc = StandardScaler()
y = y_sc.fit_transform(y).reshape(-1)
```

```python
print(X)
```

```python
print(y)
```

<h4>Training the SVR model</h4>

```python
from sklearn.svm import SVR

sv_reg = SVR(kernel = "rbf")
sv_reg.fit(X, y)
```

<h4>Visualising the SVR results</h4>

```python
import matplotlib.pyplot as plt

plt.scatter(X, y, c = "blue")
plt.plot(X, sv_reg.predict(X), c = "red")
plt.title("SVR: Position Level v Salary")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
y = y.reshape(-1, 1)

plt.scatter(X_sc.inverse_transform(X), y_sc.inverse_transform(y), c = "blue")
plt.plot(X_sc.inverse_transform(X), y_sc.inverse_transform(sv_reg.predict(X).reshape(-1, 1)), c = "red")
plt.title("SVR: Position Level v Salary")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```
